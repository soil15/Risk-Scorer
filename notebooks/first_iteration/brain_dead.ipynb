```python
def get_age(x):

    return int(str(x).split()[0])
```

```python
df['min_age'] = df['protocolSection_eligibilityModule_minimumAge'].apply(lambda x: get_age(x))
```

```python
df.drop('protocolSection_eligibilityModule_minimumAge', axis=1, inplace=True)
```

```python
df.head()
```

```python
def get_date(x):

    temp_list = x.split('-')

    if len(temp_list) != 3:
        
        temp_list.append('01')

        date = '-'.join(temp_list)

    else:
        date = x

    return date
```

```python
df['start_date'] = df['protocolSection_statusModule_startDateStruct_date'].apply(lambda x: get_date(x))

df['first_post_date'] = df['protocolSection_statusModule_studyFirstPostDateStruct_date'].apply(lambda x: get_date(x))

df[['start_date', 'first_post_date']]
```

```python
df.drop(['protocolSection_statusModule_startDateStruct_date', 'protocolSection_statusModule_studyFirstPostDateStruct_date'], axis=1, inplace=True)
```

```python
df['start_date'] = pd.to_datetime(df['start_date'])

df['first_post_date'] = pd.to_datetime(df['first_post_date'])
```

```python
df['start_date_day'] = df['start_date'].dt.day
df['start_date_month'] = df['start_date'].dt.month
df['start_date_YEAR'] = df['start_date'].dt.year

df['first_post_date_day'] = df['first_post_date'].dt.day
df['first_post_DATE_month'] = df['first_post_date'].dt.month
df['first_post_year'] = df['first_post_date'].dt.year
```

```python
df.head()
```

```python
df.drop(['start_date', 'first_post_date'], axis=1, inplace=True)
```

```python
df.head()
```

```python
to_label_encode = ['protocolSection_identificationModule_nctId', 'protocolSection_identificationModule_briefTitle', 'protocolSection_statusModule_overallStatus', 'protocolSection_sponsorCollaboratorsModule_leadSponsor_name', 'protocolSection_sponsorCollaboratorsModule_leadSponsor_class', 'protocolSection_conditionsModule_conditions_0', 'protocolSection_designModule_studyType', 'protocolSection_eligibilityModule_sex', 'protocolSection_eligibilityModule_stdAges_0', 'protocolSection_contactsLocationsModule_locations_0_city', 'protocolSection_contactsLocationsModule_locations_0_country', 'hasResults', 'protocolSection_armsInterventionsModule_interventions_0_name']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for col in to_label_encode:

    le = LabelEncoder()

    df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.metrics import precision_score, recall_score, confusion_matrix, f1_score,accuracy_score
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(dep_feature, axis=1)

y = df[dep_feature]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=15)
```

```python
rfc = RandomForestClassifier()

rfc.fit(X_train, y_train)
```

```python
y_pred = rfc.predict(X_test)
```

```python
print('Precision : {}'.format(precision_score(y_pred=y_pred, y_true=y_test)))
print('Recall : {}'.format(recall_score(y_pred=y_pred, y_true=y_test)))
print('F1_score : {}'.format(f1_score(y_pred=y_pred, y_true=y_test)))
print('Trainning Accuracy : {}'.format(rfc.score(X_train, y_train)))
print('Testing Accuracy : {}'.format(accuracy_score(y_pred=y_pred, y_true=y_test)))
```

```python
svc = SVC()

svc.fit(X_train, y_train)

y_pred = svc.predict(X_test)

print('Precision : {}'.format(precision_score(y_pred=y_pred, y_true=y_test)))
print('Recall : {}'.format(recall_score(y_pred=y_pred, y_true=y_test)))
print('F1_score : {}'.format(f1_score(y_pred=y_pred, y_true=y_test)))
print('Trainning Accuracy : {}'.format(rfc.score(X_train, y_train)))
print('Testing Accuracy : {}'.format(accuracy_score(y_pred=y_pred, y_true=y_test)))
```

```python
adc = AdaBoostClassifier()

adc.fit(X_train, y_train)

y_pred = adc.predict(X_test)

print('Precision : {}'.format(precision_score(y_pred=y_pred, y_true=y_test)))
print('Recall : {}'.format(recall_score(y_pred=y_pred, y_true=y_test)))
print('F1_score : {}'.format(f1_score(y_pred=y_pred, y_true=y_test)))
print('Trainning Accuracy : {}'.format(rfc.score(X_train, y_train)))
print('Testing Accuracy : {}'.format(accuracy_score(y_pred=y_pred, y_true=y_test)))
```

```python
xgbc = XGBClassifier()

xgbc.fit(X_train, y_train)

y_pred = xgbc.predict(X_test)

print('Precision : {}'.format(precision_score(y_pred=y_pred, y_true=y_test)))
print('Recall : {}'.format(recall_score(y_pred=y_pred, y_true=y_test)))
print('F1_score : {}'.format(f1_score(y_pred=y_pred, y_true=y_test)))
print('Trainning Accuracy : {}'.format(rfc.score(X_train, y_train)))
print('Testing Accuracy : {}'.format(accuracy_score(y_pred=y_pred, y_true=y_test)))
```
