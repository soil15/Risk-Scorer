```python
import pandas as pd
import numpy as np
import json
from flatten_json import flatten
```

### Most Popular Download

```python
df = pd.read_json('../data/most_popular.json')
```

```python
dict_0 = dict(df.iloc[0])

flat_0 = flatten(dict_0)

df_flat = pd.DataFrame(flat_0, index=[0])

for i in range(1, len(df)):
    
    dict_i = dict(df.iloc[i])

    flat_n = flatten(dict_i)

    # if i < 5:
    df_flat = pd.concat([df_flat, pd.DataFrame(flat_n, index=[0])], ignore_index=True)

df_flat
```

```python
null_df = pd.DataFrame(df_flat.isnull().sum().reset_index())

null_df.head()
```

```python
def get_perc(x):

    return (x/len(df_flat)) * 100
```

```python
null_df['null_perc'] = null_df[0].apply(lambda x : get_perc(x=x))
```

```python
perc = 80

null_df[null_df['null_perc'] > perc]
```

### Custom Download

```python
df = pd.read_json('../data/custom_download.json')
```

```python
dict_0 = dict(df.iloc[0])

flat_0 = flatten(dict_0)

df_flat = pd.DataFrame(flat_0, index=[0])

for i in range(1, len(df)):
    
    dict_i = dict(df.iloc[i])

    flat_n = flatten(dict_i)

    # if i < 5:
    df_flat = pd.concat([df_flat, pd.DataFrame(flat_n, index=[0])], ignore_index=True)

df_flat
```

```python
null_df = pd.DataFrame(df_flat.isnull().sum().reset_index())

null_df.head()
```

```python
null_df['null_perc'] = null_df[0].apply(lambda x : get_perc(x=x))
```

```python
perc = 10

null_df[null_df['null_perc'] < perc]
```

```python
final_cols = list(null_df[null_df['null_perc'] < perc]['index'])

cols_to_drop = []

for col in df_flat.columns:

    if col not in final_cols:
        cols_to_drop.append(col)

cols_to_drop
```

```python
final_df = df_flat.drop(cols_to_drop, axis=1).copy(deep=True)

final_df.head()
```

```python
final_df.to_csv('../data/final_csv.csv', index=False)
```
